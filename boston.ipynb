```python
## Depended and Independed feature
# Price is depended feature
# And remaning columns are indepedned feature
```

```python
# x is independed feature that take all excepts Price columns
x=datasets.iloc[:,:-1]
```

```python
x
```

```python
y=datasets.iloc[:,-1]
```

```python
y.head()
```

```python
#Train Testa and Splits
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
#Standradinzing the datasets
#Why?
#seed up model training
#Improves accuracy
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
```

```python
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Mode Training
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
#Print the coeffincients
print(reg.coef_)
```

```python
# ON which parameter the models has been trained
reg.get_params()
```

```python
#Predication with test data 
reg_pred=reg.predict(x_test)
```

```python
reg_pred
```

ASSUMPTIONS

```python
# Plot a scatter plot for the predications
plt.scatter(reg_pred,y_test)
plt.show()
```

```python
#Residulas values ->error -> the vallue different between obsesrved value and calulated values
```

```python
residuals=y_test-reg_pred
```

```python
residuals
```

```python
# Plot's the residulas
sns.displot(residuals,kind='kde')
plt.show()
```

```python
# Scatter plot's with respects to predications and residuals
# Uniforms distributions
plt.scatter(reg_pred,residuals)
```

```python
# Now we are going to used performanced matrix
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse=mean_squared_error(reg_pred,residuals)
mae=mean_absolute_error(reg_pred,residuals)
```

```python
mse
```

```python
mae
```

R square and Adjusted R squared

FORMULA
R^2=1-SSR/SST

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
```
